```python
X = np.random.rand(2, 100)
A = np.random.rand(2,2)
e, v = np.linalg.eig(A)
print(v)
plt.figure()
plt.scatter(X[0, :], X[1, :])
plt.title("Scatter Points")
plt.xlabel("X")
plt.ylabel("Y")
center = np.mean(X, axis = 1)
print(center)
plt.arrow(center[0], center[0],v[0, 0], v[0, 1])
plt.arrow(center[0], center[0],v[1, 0], v[1, 1])
plt.axis('equal')
```
